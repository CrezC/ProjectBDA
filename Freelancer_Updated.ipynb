## Class Overview
The `FreelancerAnalysis` class handles:

- Filtering and analysis
- Visualization of contract trends


### Data Loading and Parsing

```python

from freelancer_analysis_updated import FreelancerAnalysis

# Initialize analysis
analysis = FreelancerAnalysis('data/contracts.csv')
```

### Filtering and Analysis

1. `plot_monthly_timeline`: Shows contract volume over time
